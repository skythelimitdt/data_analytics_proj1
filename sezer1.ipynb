```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import hvplot.pandas
```

```python
# Take in all of our data and read it into pandas

world_energy_df = pd.read_csv("Resources/world_energy_consumption.csv")
world_energy_df.head()
```

```python
world_energy_df.columns.tolist()
```

```python
#Look at the data type for each of the column
pd.set_option("display.max_columns", None)
world_energy_df.info(verbose=True)
```

```python
# Display all rows with at least 1 missing data (WE CAN DELETE)
combined_conditions = world_energy_df[
                    world_energy_df.isna().any(axis=1) | (world_energy_df == '').any(axis=1) | (world_energy_df == 0).any(axis=1)]
combined_conditions.head()
```

```python
# Display the data count on each column in our data set
pd.set_option("display.max_columns", None)
count_result = world_energy_df.count()
count_result
```

```python
#Remove all rows that does not have any data which indicates those rows do not represent the countries
countries_world_energy_df=world_energy_df.dropna(subset=['iso_code'])
countries_world_energy_df
```

```python
filtered_countries=countries_world_energy_df.loc[(countries_world_energy_df["year"] >= 1970)] 
filtered_countries.head()
```

```python
# Display the 
pd.set_option("display.max_columns", None)
count_result = countries_world_energy_df.count()
count_result
```

```python
six_continents = world_energy_df.loc[
    (
        (world_energy_df["country"] == "Asia") |
        (world_energy_df["country"] == "Africa") |
        (world_energy_df["country"] == "Australia") |
        (world_energy_df["country"] == "Europe") |
        (world_energy_df["country"] == "North America") |
        (world_energy_df["country"] == "South America")
    ) & (world_energy_df["year"] >= 1970)  
]
six_continents
```

```python
# Plot energy consumption over time by continent
consumption_plot_1a = six_continents.hvplot.line(
    x='year', y='coal_consumption', by='country', 
    title="Coal Consumption Over Time by Continent",
    xlabel='Year', ylabel='Coal Consumption',
    line_width=2, height=400, width=1000
)
consumption_plot_1a
```

```python
# Plot energy production over time by continent
consumption_plot_1b = six_continents.hvplot.line(
    x='year', y='coal_production', by='country', 
    title="Coal Production Over Time by Continent",
    xlabel='Year', ylabel='Coal Production',
    line_width=2, height=400, width=1000
)
consumption_plot_1b
```

```python
# Filter data for the year 2020
year_2022_data = six_continents[six_continents['year'] == 2022]

# Plot coal consumption for 2022 by continent
consumption_2022_plot = year_2022_data.hvplot.bar(
    x='country', y='coal_consumption', 
    title="Coal Consumption by Continent in 2022",
    xlabel='Continent', ylabel='Coal Consumption',
    color='blue', height=400, width=600
)

# Plot coal production for 2020 by continent
production_2022_plot = year_2022_data.hvplot.bar(
    x='country', y='coal_production', 
    title="Coal Production by Continent in 2022",
    xlabel='Continent', ylabel='Energy Production',
    color='green', height=400, width=600
)

(consumption_2022_plot + production_2022_plot).cols(1)
```

```python
# Example of energy mix if renewables, coal, oil columns exist
energy_mix_plot = year_2022_data.hvplot.bar(
    x='country', y=['renewables_consumption', 'coal_consumption', 'oil_consumption', 'solar_consumption','nuclear_consumption','gas_consumption'], 
    stacked=True, title="Energy Mix by Continent in 2022",
    xlabel='Continent', ylabel='Energy Source Contribution',
    height=400, width=800
)
energy_mix_plot
```

```python
#"iso_code", "population", "gdp"
renewable_energy_df=six_continents.loc[:,["country", "year", "renewables_consumption", "solar_consumption", "biofuel_consumption", "hydro_consumption", "wind_consumption","other_renewable_consumption"]]
renewable_energy_df
```

```python
non_renewable_energy_df=six_continents.loc[:,["country", "year", "coal_consumption", "fossil_fuel_consumption","gas_consumption","low_carbon_consumption","nuclear_consumption","oil_consumption"]]
non_renewable_energy_df
```

```python
# Sum of renewable energy consumption by country and year
renewable_energy_df['total_renewable'] = renewable_energy_df[
    ["renewables_consumption", "solar_consumption", "biofuel_consumption", "hydro_consumption", "wind_consumption", "other_renewable_consumption"]
].sum(axis=1)
# Plot renewable energy consumption over time
renewable_plot = renewable_energy_df.hvplot.line(
    x='year', y='total_renewable', by='country', 
    title="Total Renewable Energy Consumption Over Time by Country",
    xlabel='Year', ylabel='Total Renewable Energy Consumption',
    line_width=2, height=500, width=800
)
renewable_plot
```

```python
# Sum of non-renewable energy consumption by country and year
non_renewable_energy_df['total_non_renewable'] = non_renewable_energy_df[
    ["coal_consumption", "fossil_fuel_consumption", "gas_consumption", "low_carbon_consumption", "nuclear_consumption", "oil_consumption"]
].sum(axis=1)

# Plot non-renewable energy consumption over time
non_renewable_plot = non_renewable_energy_df.hvplot.line(
    x='year', y='total_non_renewable', by='country', 
    title="Total Non-Renewable Energy Consumption Over Time by Country",
    xlabel='Year', ylabel='Total Non-Renewable Energy Consumption',
    line_width=2, height=400, width=800
)
non_renewable_plot
```

```python
# Overlay renewable and non-renewable plots for comparison
combined_plot = renewable_plot * non_renewable_plot
combined_plot.opts(title="Renewable vs Non-Renewable Energy Consumption Over Time by Country", ylabel="Consumption", border_line_width=2, height=800, width=1200)
combined_plot 
```

```python
# Add or rename the total renewable and non-renewable energy columns
renewable_energy_df['total_energy'] = renewable_energy_df[
    ["renewables_consumption", "solar_consumption", "biofuel_consumption", 
     "hydro_consumption", "wind_consumption", "other_renewable_consumption"]
].sum(axis=1)

non_renewable_energy_df['total_energy'] = non_renewable_energy_df[
    ["coal_consumption", "fossil_fuel_consumption", "gas_consumption", 
     "low_carbon_consumption", "nuclear_consumption", "oil_consumption"]
].sum(axis=1)
# Label data as renewable or non-renewable for each DataFrame
renewable_energy_df['type'] = 'Renewable'
non_renewable_energy_df['type'] = 'Non-Renewable'

# Combine both DataFrames into one
combined_df = pd.concat([renewable_energy_df[['year', 'country', 'total_energy', 'type']],
                         non_renewable_energy_df[['year', 'country', 'total_energy', 'type']]])

# Confirm combined DataFrame columns
print(combined_df.columns)

```

```python
# Plot with stacked bars for renewable and non-renewable energy
combined_stacked_bar = combined_df.hvplot.bar(
    x='year', y='total_energy', by='type', stacked=True, 
    title="Renewable vs Non-Renewable Energy Consumption by Country",
    xlabel='Year', ylabel="Total Energy Consumption", height=800, width=2000
)
combined_stacked_bar


```
